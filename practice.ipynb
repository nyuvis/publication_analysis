```python
import numpy as np
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("./IEEE VIS papers 1990-2018 - Main dataset.csv")
df.head(10)
```

## remove nan values in abstract
remove rows if abstract is empty, before: 3109 rows, after: 3035 rows

```python
df = df.dropna(subset=['Abstract'])
```

```python
# number of abstracts presented in different years

years = list(df["Year"].unique())
freq = list(df["Year"].value_counts(sort=False))

to_plot = pd.DataFrame()
to_plot["Years"] = years
to_plot["Number of Abstracts"] = freq

alt.Chart(to_plot).mark_line().encode(
    x = "Years:Q",
    y = "Number of Abstracts:Q"
)
```

```python
## Revised version

to_plot1 = df.groupby('Year').count().reset_index()
alt.Chart(to_plot1).mark_line().encode(
    x = "Year:Q",
    y = "Title:Q"
)
```

```python
to_plot1
```

```python
# total papers in each track 

tracks = list(df["Conference"].unique())
vis = (df.Conference == 'Vis').sum()
infovis = (df.Conference == 'InfoVis').sum()
vast = (df.Conference == 'VAST').sum()
scivis = (df.Conference == 'SciVis').sum()
numPapers = [vis, infovis, vast, scivis]

to_plot2 = pd.DataFrame()
to_plot2['Tracks'] = tracks
to_plot2['Total Papers'] = numPapers

alt.Chart(to_plot2).mark_bar().encode(
    x = "Tracks:N",
    y = "Total Papers:Q"
)
```

```python
to_plot2 = df.groupby('Conference').count().reset_index()

alt.Chart(to_plot2).mark_bar().encode(
    x = "Conference:N",
    y = "Abstract:Q",
)
```

```python
to_plot2
```

```python
# number of abstracts accepted by different tracks in different years
to_plot3 = df.groupby(['Conference', 'Year']).count().reset_index()
```

```python
alt.Chart(to_plot3).mark_line().encode(
    x = "Year:Q",
    y = "Title:Q",
    color = 'Conference:N'
)
```
